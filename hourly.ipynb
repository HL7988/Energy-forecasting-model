```python
hourly = pd.read_csv('hourly.csv',skiprows=2)
```

```python
hourly['Sum'] = hourly.iloc[:,1:349].sum(axis=1)
```

```python
df_copy = pd.DataFrame().reindex(columns=hourly.columns)
```

```python
df_copy
```

```python
i=0
while(i<52560):
    df_copy.loc[i/6,1:350]=hourly.iloc[i:i+6,1:350].sum(axis=0)
    df_copy.loc[i/6,'daytime']=hourly.iloc[i,0]
    i+=6
```

```python
df_copy
```

```python
dict = {'Day':[],
        'Month':[],
        'Year':[],
        'time':[],
       }
df_new = pd.DataFrame(dict)
df_new
```

```python
i=0
while(i<8760):
    date,time=df_copy['daytime'][i].split()
    month,day,year=date.split('/')
    df_new.loc[i]=[day,month,year,time]
    i+=1
```

```python
df_new['Power']=df_copy['Sum']
df_new['Datetime']=df_copy['daytime']
df_new
```

```python
df_new.to_csv('reduce_hourly.csv',index=False)
```
